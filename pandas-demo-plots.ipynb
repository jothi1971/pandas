```python
#reference - https://pandas.pydata.org/docs/getting_started/intro_tutorials/04_plotting.html
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#use datetime as index. consider datetime as timestamp format and not as string
#index column is used as x-axis in the plot
#csv file has air quality data at different stations
air_quality = pd.read_csv("air_quality.csv", index_col=0, parse_dates=True)
```

```python
#display top 5 rows
air_quality.head()
```

```python
#plot the data. one plot for each column with numeric data
air_quality.plot()
```

```python
#plot with x and y range limit
axis = air_quality["2019-05-08":"2019-06-01"].plot()
axis.set_ylim(0,80)
```

```python
#plot data for paris only. chang the color
air_quality["station_paris"].plot(color="r")
```

```python
# plot types - ['area','bar','barh','box','density','hexbin','hist','kde','line','pie','scatter']
#scatter plot prints one dot for each observation
# 2 columns are of same length
# alpha - 0 transparent and 1 opaque
# compare london and paris data
air_quality.plot.scatter(x="station_london", y="station_paris", alpha=0.5)
```

```python
#use boxplot method from plot
air_quality.plot.box()
```

```python
#use of suplot argment. seperate sbplot for each column
axs = air_quality.plot.area(figsize=(12, 4), subplots=True)
```

```python
#create empty fig
#customise plot. change y label and save plot in a image file
fig, axs = plt.subplots(figsize=(12, 4))
#save the plot to the empty figure
air_quality.plot.area(ax=axs)
axs.set_ylabel("NO$_2$ concentration")
fig.savefig("air_concentrations.png")
```
